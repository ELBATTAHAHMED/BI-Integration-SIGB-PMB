Clean Data

```python
import pandas as pd
```

📥 Étape 1 : Charger les fichiers

```python
buf = pd.read_csv("FixedData/buf_fixed.csv")
bua = pd.read_csv("FixedData/bua_fixed.csv")

print("🔎 Données BUF :")
print(buf.head(3))
print("🔎 Données BUA :")
print(bua.head(3))
```

🧽 Étape 2 : Supprimer les doublons

```python
buf = buf.drop_duplicates()
bua = bua.drop_duplicates()

print(f"✅ BUF après suppression des doublons : {buf.shape[0]} lignes")
print(f"✅ BUA après suppression des doublons : {bua.shape[0]} lignes")
```

🧼 Étape 3 : Nettoyer les noms de colonnes

```python
buf.columns = buf.columns.str.strip()
bua.columns = bua.columns.str.strip()

print("🧪 Colonnes BUF :", buf.columns.tolist())
print("🧪 Colonnes BUA :", bua.columns.tolist())
```

✂️ Étape 4 : Supprimer les guillemets dans la colonne Cote

```python
buf["Cote"] = buf["Cote"].astype(str).str.replace('"', '', regex=False)
bua["Cote"] = bua["Cote"].astype(str).str.replace('"', '', regex=False)

print("✅ Extrait 'Cote' BUF :", buf["Cote"].head(3).tolist())
print("✅ Extrait 'Cote' BUA :", bua["Cote"].head(3).tolist())
```

🔢 Étape 5 : Convertir les colonnes Annee et Nb pages en numériques

```python
buf["Annee"] = pd.to_numeric(buf["Annee"], errors="coerce")
buf["Nb pages"] = pd.to_numeric(buf["Nb pages"], errors="coerce")

bua["Annee"] = pd.to_numeric(bua["Annee"], errors="coerce")
bua["Nb pages"] = pd.to_numeric(bua["Nb pages"], errors="coerce")

print("📋 BUF types :")
print(buf[["Annee", "Nb pages"]].dtypes)
print("📋 BUA types :")
print(bua[["Annee", "Nb pages"]].dtypes)
```

Vérifier les valeurs manquantes 

```python
print("❗ Valeurs manquantes BUF :")
print(buf.isnull().sum())
print("❗ Valeurs manquantes BUA :")
print(bua.isnull().sum())
```

🧮 Étape 6 : Générer un inventaire unique (si vide ou manquant)

```python
buf["Inventaire"] = buf["Inventaire"].fillna("INV_BUF")
buf["Inventaire"] = ["INV_BUF_" + str(i) for i in range(1, len(buf) + 1)]

bua["Inventaire"] = bua["Inventaire"].fillna("INV_BUA")
bua["Inventaire"] = ["INV_BUA_" + str(i) for i in range(1, len(bua) + 1)]

print("🆔 Exemple d'inventaires BUF :", buf["Inventaire"].head(3).tolist())
print("🆔 Exemple d'inventaires BUA :", bua["Inventaire"].head(3).tolist())
```

⚠️ Étape 7 : Supprimer les lignes essentielles vides (Titre ou Inventaire)

```python
bua = bua.dropna(subset=["Titre", "Inventaire"])

print(f"✅ BUA lignes après suppression des titres/inventaires manquants : {bua.shape[0]}")
```

🧩 Étape 8 : Gérer les valeurs manquantes dans buf (français)

```python
buf["Auteur"] = buf["Auteur"].fillna("Inconnu")
buf["Lieu"] = buf["Lieu"].fillna("Inconnu")
buf["Edition"] = buf["Edition"].fillna("Non spécifié")
buf["Matiere"] = buf["Matiere"].fillna("Inconnu")
buf["Matiere"] = buf["Matiere"].str.replace("\xa0", " ").str.strip()


buf["Annee"] = buf["Annee"].fillna(buf["Annee"].median())
buf["Nb pages"] = buf["Nb pages"].fillna(buf["Nb pages"].median())

print("📊 Vérification finale BUF :")
print(buf.isnull().sum())
```

🧩 Étape 9 : Gérer les valeurs manquantes dans bua (arabe)

```python
bua["Lieu"] = bua["Lieu"].fillna("غير محدد")
bua["Edition"] = bua["Edition"].fillna("غير محدد")
bua["Matiere"] = bua["Matiere"].fillna("غير محدد")
bua["Matiere"] = bua["Matiere"].str.replace("\xa0", " ").str.strip()

bua["Annee"] = bua["Annee"].fillna(bua["Annee"].median())
bua["Nb pages"] = bua["Nb pages"].fillna(bua["Nb pages"].median())
```

💾 Étape 10 : Sauvegarder les fichiers nettoyés

```python
import os
os.makedirs("CleanedData", exist_ok=True)

buf.to_csv("CleanedData/cleaned_buf.csv", index=False, encoding="utf-8")
bua.to_csv("CleanedData/cleaned_bua.csv", index=False, encoding="utf-8")

print("💾 Fichiers nettoyés enregistrés dans CleanedData/")
```
