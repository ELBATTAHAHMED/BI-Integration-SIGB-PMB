```python
import pandas as pd
```

# === Étape 1 : Convertir bua.xls (avec headers déjà existants) ===
# Lecture du fichier BUA (en arabe)

```python
bua = pd.read_excel("FirstData/bua.xls") 
```

# Sauvegarde directe au format CSV 


```python
bua.to_csv("FixedData/bua_fixed.csv", index=False, encoding="utf-8")
print("✅ bua.xls a été converti en bua_fixed.csv avec les headers existants")
```

# === Étape 2 : Ajouter les headers à buf.csv (qui n’en a pas) ===
# Définition des noms de colonnes (assure-toi que ces noms sont corrects pour ton fichier BUF)

```python
headers = ["Cote", "Titre", "Auteur", "Lieu", "Edition", "Annee", "Nb pages", "Matiere", "Inventaire"]
```

# Lecture du fichier BUF sans en-tête

```python
buf = pd.read_csv("FirstData/buf.csv", header=None, encoding="ISO-8859-1", delimiter=";") 
```

# Ajout des noms de colonnes

```python
buf.columns = headers
```

# Sauvegarde du fichier BUF corrigé avec en-têtes

```python
buf.to_csv("FixedData/buf_fixed.csv", index=False, encoding="utf-8")
print("✅ buf.csv a été transformé en buf_fixed.csv avec en-têtes ajoutés")
```
